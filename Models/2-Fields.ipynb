## How to add enumeration fields in Database table with Django?

In the model class define a enumeration class with respected attributes / dropdown fields.

```py
class Post(models.Model):
    class Status(models.TextChoices):
        DRAFT = 'DF', 'Draft'
        PUBLISHED = 'PB','Published'

    ...
    status = models.CharField(
        max_length=2,
        choices=Status,
        default=Status.DRAFT 
    )
```
